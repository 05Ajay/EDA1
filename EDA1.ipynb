```python
#load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Tasks:
# 1.Data Cleaning and Preparation:
# Load the dataset into a DataFrame or equivalent data structure.
cardio=pd.read_csv("Cardiotocographic.csv")
cardio
```

```python
#Dropping Duplicate Records
cardio.drop_duplicates(inplace=True)
```

```python
cardio.duplicated().sum()
```

```python
# % of missing values
for i in cardio.isna().sum():
 print((i/len(cardio))*100)
  
# 0 to 5 % missing values: drop rows
# 6 to 45 %: replace the missing values
# above 50%: drop column
```

```python
cardio.dropna(inplace = True)
```

```python
cardio.isna().sum()
#missing values removed
```

```python
# Detect and treat outliers if necessary.
cardio.boxplot()
```

```python
# Iterate through each column in the dataset
for i in cardio.columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = cardio[i].quantile(0.25)
    Q3 = cardio[i].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with the respective bounds
    cardio.loc[cardio[i] < lower_bound, i] = lower_bound
    cardio.loc[cardio[i] > upper_bound, i] = upper_bound
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=cardio)
# Display the plot
plt.show()

# 1.LB (Left Bundle): 
#    This has a relatively high median value (around 140) compared to the other metrics.
#    The range (from the box and whiskers) indicates some variability, but it's generally more stable.

# 2. AC, FM, UC, DL, DS, DP:
#     These metrics show very low or near-zero median values.
#     There is little to no variability (the boxes are compressed to lines, and whiskers are almost non-existent).

# 3. ASTV (Acceleration of ST segment variance):
#     Has a moderate median value, with some variability as shown by the size of the box and whiskers.

# 4. MSTV (Mean of ST segment variance) and ALTV (Acceleration of Long-Term Variance):
#     Both have small medians but with noticeable variability, suggesting some fluctuation in these values across different samples.

# 5. MLTV (Mean of Long-Term Variance):
#     Similar to ASTV, this shows a small median with some variability.

# 6. Width Tendency:
#     This metric has a significant range, with values extending from negative to very high positive values.
#     The median is positive, but the large range indicates a wide variability in this metric.

# 7. NSP:
#     Similar to AC, FM, UC, DL, DS, and DP, this metric shows a near-zero median with almost no variability.
```

```python
# Visualization for LB - Baseline Fetal Heart Rate (FHR) using Histogram(seaborn)
plt.figure(figsize=(10, 6))
sns.histplot(cardio['LB'], bins=20, kde=True)
plt.title('Histogram of Baseline Fetal Heart Rate (FHR)')
plt.xlabel('Baseline Fetal Heart Rate (FHR)')
plt.ylabel('Frequency')
plt.show()

# A baseline FHR between 110 and 160 bpm is generally considered normal during fetal monitoring.
# The histogram shows that the data is concentrated within this range,
# indicating that the majority of the readings fall within the normal range.
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation = cardio.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()

# - Strong Relationships:
#   - The strong positive correlation between MSTV and MLT suggests that these two variables tend to increase or decrease together.
#   - The negative correlations between ALTV with ASTV and MSTV could indicate that as ALTV increases, ASTV and MSTV decrease, and vice versa.

# - Independent Variables:
#   - Some variables like LB and FM have little to no correlation with the majority of other metrics, suggesting that they may operate independently of the other variables in this dataset.
```

```python
# Creating violin plots for each variable
vars_to_plot = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']
plt.figure(figsize=(12, 8))
for i, var in enumerate(vars_to_plot):
    plt.subplot(4, 3, i + 1)
    sns.violinplot(x=cardio[var])
    plt.title(var)
plt.tight_layout()
plt.show()
```
